```python
# Generate data
X, _ = make_blobs(
    n_samples=1000, centers=3, n_features=3, random_state=SEED)

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test = train_test_split(
    X_scaled, test_size=0.2, random_state=SEED)
```

####ii. Побудувати двi моделi лiнiйного автокодувальника: одну - для двовимiрного, iншу - для одновимiрного представлення.

Напишемо клас для створення простих автокодувальників із лінійними або нелінійними функціями активації для представлення даних в одному або 2 вимірах. Для навчання застосовуватимемо функцію втрат MSE та оптимізатор Adam. Кількість епох для навчання встановимо рівну 100 за засовчуванням.

```python
class AutoEncoder:
  def __init__(self, input_size=3, hidden_size=2, activation=None):
    self._input_size = input_size
    self._hidden_size = hidden_size
    self._output_size = self._input_size
    self._activation = activation

    inputs = keras.Input(shape=(self._input_size,))

    # Encoder
    encoded = keras.layers.Dense(
        self._hidden_size, activation=self._activation)(inputs)
    self._encoder = keras.Model(inputs, encoded)

    # Decoder
    decoded = keras.layers.Dense(
        self._output_size, activation=self._activation)(encoded)

    self._model = keras.Model(inputs, decoded)

  def compile(self, optimizer="adam", loss="mse"):
    self._model.compile(optimizer=optimizer, loss=loss)

  def fit(self, X_train, epochs=100, batch_size=BATCH_SIZE, verbose=0):
    self._model.fit(X_train, X_train, epochs=epochs,
                    batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, verbose=0):
    return self._model.evaluate(X_test, X_test, verbose=verbose)

  def encode(self, X_test):
    return self._encoder.predict(X_test)
```

Створимо лінійні автокодувальники для представлення даних у двовимірному та одновимірному просторах.

```python
# 2D representation (linear autoencoder)
linear_autoencoder_2d = AutoEncoder()
linear_autoencoder_2d.compile()

# 1D representation (linear autoencoder)
linear_autoencoder_1d = AutoEncoder(hidden_size=1)
linear_autoencoder_1d.compile()
```

###2. Побудувати глибокi моделi шумопригнiчуючого i регуляризованого автокодувальника для реконструкцiї зображень

Створимо клас для шумопригнічуючого автококдувальника з використанням гаусівського шуму та шарів Dropout. Викоистовуватимемо активації ReLU всередині автокодувальника та Sigmoid на його виході. Автокодувальник матиме симетричну структуру, розміри шарів: 256, 128, 64, 128, 256. Для тренування використовуватимемо оптимізатор Adam і функцію втрат MSE.

```python
BATCH_SIZE = 128

class DenoisingAutoEncoder:
  def __init__(self, input_size=784, hidden_sizes=[256, 128, 64],
               noise_std=0.1, dropout_rate=0.0, activation="relu"):
    self._input_size = input_size
    self._hidden_sizes = hidden_sizes
    self._output_size = self._input_size
    self._noise_std = noise_std
    self._dropout_rate = dropout_rate
    self._activation = activation

    inputs = keras.Input(shape=(self._input_size,))
    x = inputs

    # Gaussian noise
    if self._noise_std > 0.0:
      x = keras.layers.GaussianNoise(self._noise_std)(x)

    # Dropout
    if self._dropout_rate > 0.0:
      x = keras.layers.Dropout(self._dropout_rate)(x)

    # Encoder
    for size in self._hidden_sizes[:-1]:
      x = keras.layers.Dense(size, activation=self._activation)(x)

    encoded = keras.layers.Dense(
        self._hidden_sizes[-1], activation=self._activation)(x)

    self._encoder = keras.Model(inputs, encoded)

    # Decoder (symmetric)
    x = encoded
    for size in reversed(hidden_sizes[:-1]):
      x = keras.layers.Dense(size, activation=self._activation)(x)

    outputs = keras.layers.Dense(
        self._output_size, activation="sigmoid")(x)

    self._model = keras.Model(inputs, outputs)

  def compile(self, optimizer="adam", loss="mse"):
    self._model.compile(optimizer=optimizer, loss=loss)

  def fit(self, X_train, X_val, epochs=10, batch_size=BATCH_SIZE, verbose=1):
    self._model.fit(X_train, X_train, epochs=epochs,
                    validation_data=(X_val, X_val),
                    batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, verbose=1):
    return self._model.evaluate(X_test, X_test, verbose=verbose)

  def encode(self, X_test):
    return self._encoder.predict(X_test)

  def reconstruct(self, X_test):
    return self._model.predict(X_test)
```

###3. Виконати попереднє навчання класифiкатора зображень, використовуючи шумопригнiчуючий автокодувальник

####i. Завантажити обраний набiр даних iз зображеннями.

Завнатажимо набір даних Fashion MNIST та виконаємо його попередню обробку.

```python
(X_full_train, y_full_train), (X_test, y_test) = \
tf.keras.datasets.fashion_mnist.load_data()

X_full_train = X_full_train.astype(np.float32).reshape(-1, 28*28) / 255.0
X_test = X_test.astype(np.float32).reshape(-1, 28*28) / 255.0
y_full_train = y_full_train.astype(np.int32)
y_test = y_test.astype(np.int32)
```

####ii. Розбити данi на три пiдмножини: навчальну; валiдацiйну - для налаштування значень гiперпараметрiв; тестову - для остаточної оцiнки якостi.

Розіб'ємо данi на три пiдмножини: навчальну, валiдацiйну та тестову.

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X_full_train, y_full_train, test_size=0.1, random_state=SEED)
```

####iii. Побудувати глибоку модель шумопригнiчуючого автокодувальника. Навчити її на навчальному наборi.

Побудуємо модель шумопригнічуючого автокодувальника та навчимо її на даних Fashion MNIST.

```python
denoising_autoencoder = DenoisingAutoEncoder()
denoising_autoencoder.compile()
denoising_autoencoder.fit(X_train, X_valid)
```

####iv. Спробувати пiдiбрати гiперпараметри моделi, використовуючи перехресну перевiрку, так щоб модель гарно виконувала реконструкцiю зображень.

Підберемо найкращі параметри для моделі шумопригнічуючого автокодувальника з використанням перехресної перевірки. Розглядатимемо різні значення параметрів dropout_rate (0.0, 0.1, 0.3) та noise_std (0.0, 0.1, 0.3). Для перехресної перевірки використаємо 3 фолди.

```python
noise_stds = [0.0, 0.1, 0.3]
dropout_rates = [0.0, 0.1, 0.3]
best_loss = float("inf")
best_params = None

kf = KFold(n_splits=3, shuffle=True, random_state=SEED)

for noise_std in noise_stds:
  for dropout_rate in dropout_rates:
    val_losses = []

    for train_index, val_index in kf.split(X_train):
      X_tr, X_val = X_full_train[train_index], X_full_train[val_index]

      model = DenoisingAutoEncoder(
          noise_std=noise_std, dropout_rate=dropout_rate)
      model.compile()
      model.fit(X_tr, X_val, epochs=5, verbose=0)

      loss = model.evaluate(X_val)
      val_losses.append(loss)

    avg_loss = np.mean(val_losses)
    print(f"Tested noise={noise_std}, dropout={dropout_rate}:",
          f"val MSE: {avg_loss:.4f}")

    if avg_loss < best_loss:
      best_loss = avg_loss
      best_params = (noise_std, dropout_rate)

print(f"Best hyperparameters: {best_params} with val loss: {best_loss:.4f}")
```

Бачимо, що найкраща комбінація гіперпараметрів для шумопригнічуючого автокодувальника - це noise_std=0.1 та dropout_rate=0.0.

* На основi найнижчих шарів моделі побудувати глибоку нейронну мережу для класифiкацiї.

Створимо клас класифікатора, що працює на базі кодувальника шумопригнічуючого автокодувальника. Він матиме два приховані шари (рзмірів 32 і 16) та вихідний шар. У прихованих шарах використовується активація ReLU, а у вихідному - Softmax. Під час навчання навчатимемо лише класифікуючу голову - шари енкодера змінювати не будемо. Навчання виконуватиметься оптимізатором Adam з функцією втрат Sparse Categorical Crossentropy та обчисленням метрики Accuracy.

Також створимо модель цього класу та скомпілюємо її.

```python
class Classifier:
  def __init__(self, autoencoder, hidden_sizes=[32, 16], output_size=10,
               activation="relu"):
    self._autoencoder = autoencoder
    self._hidden_sizes = hidden_sizes
    self._output_size = output_size
    self._activation = activation

    for layer in self._autoencoder._encoder.layers:
      layer.trainable = False

    x = self._autoencoder._encoder.output

    for size in self._hidden_sizes:
      x = keras.layers.Dense(size, activation=self._activation)(x)

    x = keras.layers.Dense(self._output_size, activation="softmax")(x)

    self._classifier = keras.Model(self._autoencoder._encoder.input, x)

  def compile(self, optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]):
    self._classifier.compile(
        optimizer=optimizer, loss=loss, metrics=metrics)

  def fit(self, X_train, y_train, X_val, y_val, epochs=10,
          batch_size=BATCH_SIZE, verbose=1):
    self._classifier.fit(
        X_train, y_train, epochs=epochs,
        validation_data=(X_val, y_val),
        batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, y_test, verbose=1):
    return self._classifier.evaluate(X_test, y_test, verbose=verbose)

small_classifier = Classifier(denoising_autoencoder)
small_classifier.compile()
```

###4. Сгенерувати новi зображення, використовуючи варiацiйний автокодувальник

####i. Побудувати модель варiацiйного автокодувальника. Функцiю втрат визначити як суму втрати реконструкцiї та латентної втрати.

Побудуємо модель варіаційного автокодувальника. Розміри його внутрішніх шарів кодувальника (і декодувальника, в оберненому порядку) становитимуть 512, 256. Розмір латентного простору - 20. Використовуватимемо функцію активації ELU. Втрату обчислюватимемо як суму втрати реконструкції (Sigmoid Cross Entropy) та модифікованої латентної втрати.

```python
class VariationalAutoEncoder(keras.Model):
  def __init__(self, input_size=784, hidden_sizes=[512, 256],
               latent_size=20, activation="elu"):
    super().__init__()
    self._input_size = input_size
    self._hidden_sizes = hidden_sizes
    self._latent_size = latent_size
    self._output_size = self._input_size
    self._activation = activation
    self._eps = 1e-10

    inputs = keras.Input(shape=(self._input_size,))
    x = inputs

    # Encoder
    for size in self._hidden_sizes:
      x = keras.layers.Dense(size, activation=self._activation)(x)

    z_mean = keras.layers.Dense(self._latent_size)(x)
    z_log_var = keras.layers.Dense(self._latent_size)(x)

    def sampling(args):
      z_mean, z_log_var = args
      epsilon = tf.random.normal(shape=tf.shape(z_mean))
      return z_mean + tf.exp(0.5 * z_log_var) * epsilon

    z = keras.layers.Lambda(sampling)([z_mean, z_log_var])
    self._encoder = keras.Model(inputs, [z_mean, z_log_var, z])

    # Decoder
    latent_inputs = keras.Input(shape=(self._latent_size,))
    x = latent_inputs
    for size in reversed(self._hidden_sizes):
      x = keras.layers.Dense(size, activation=self._activation)(x)

    logits = keras.layers.Dense(self._output_size)(x)
    self._decoder = keras.Model(latent_inputs, logits)

  def call(self, inputs):
    z_mean, z_log_var, z = self._encoder(inputs)
    logits = self._decoder(z)

    # Reconstruction loss
    reconstruction_loss = tf.reduce_sum(
        tf.nn.sigmoid_cross_entropy_with_logits(labels=inputs, logits=logits),
        axis=1
    )

    # Latent loss
    std = tf.exp(0.5 * z_log_var)
    kl_loss = 0.5 * tf.reduce_sum(
        tf.square(std) + tf.square(z_mean) - 1 - tf.math.log(self._eps + tf.square(std)),
        axis=1
    )

    total_loss = tf.reduce_mean(reconstruction_loss + kl_loss)
    self.add_loss(total_loss)

    return tf.nn.sigmoid(logits)
```

###5. Виконати семантичне хешування зображень iз застосуванням модифiкованого глибокого автокодувальника

####i. Створити першу модель глибокого автокодувальника, яка мiстить два скритих шари i наступний за ними кодуючий скритий шар. Розглянути декiлька варiантiв кодувальних шарiв з 20, 30 i 40 нейронами (iншою кiлькiстю?) i логiстичною функцiєю активацiї.

Створимо клас глибокого автокодувальника для семантичного хешування зображень. Розміри його внутрішніх шарів кодувальника (і декодувальника, в оберненому порядку) становитимуть 512, 256. Розмір кодуючого шару - 30. Використовуватимемо функцію активації ReLU для прихованих шарів, а для вихідного та кодуючого - Sigmoid. Втрату обчислюватимемо за допомогою Binary Crossentropy. Застосовуватимемо оптимізатор Adam.

```python
class HashAutoEncoder:
  def __init__(self, input_size=784, hidden_sizes = [512, 256],
               code_size=30, activation="relu"):
    self._input_size = input_size
    self._hidden_sizes = hidden_sizes
    self._output_size = self._input_size
    self._code_size = code_size
    self._activation = activation

    inputs = keras.Input(shape=(self._input_size,))
    x = inputs
    # Encoder
    for size in self._hidden_sizes:
      x = keras.layers.Dense(size, activation=self._activation)(inputs)

    code = keras.layers.Dense(self._code_size, activation="sigmoid")(x)
    self._encoder = keras.Model(inputs, code)

    x = code
    # Decoder
    for size in reversed(self._hidden_sizes):
      x = keras.layers.Dense(size, activation=self._activation)(x)

    outputs = keras.layers.Dense(self._output_size, activation="sigmoid")(x)

    self._model = keras.Model(inputs, outputs)

  def compile(self, optimizer="adam", loss="binary_crossentropy"):
    self._model.compile(optimizer=optimizer, loss=loss)

  def fit(self, X_train, X_val, epochs=10, batch_size=BATCH_SIZE,
          verbose=1):
    self._model.fit(
        X_train, X_train, epochs=epochs,
        validation_data=(X_val, X_val),
        batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, verbose=1):
    return self._model.evaluate(X_test, X_test, verbose=verbose)

  def encode(self, X_test):
    return self._encoder.predict(X_test)

  def reconstruct(self, X_test):
    return self._model.predict(X_test)

  def encode_hash(self, X_test):
    codes = self.encode(X_test)
    return (codes > 0.5).astype(int)
```

####iv. Створити другу модель шляхом додавання гаусiвського шуму з нульовим середнiм до входiв кодуючого шару першої моделi тiльки пiд час навчання. Для того щоб вiдношення сигналу до шуму зберiгалося високим, автокодувальник буде вчитися надавати великi значення кодуючому шару i таким чином шум стає незначним. Тому логiстична функцiя кодуючого шару, скорiш за все, буде насичуватися в точцi 0 або 1.

Створимо клас глибокого автокодувальника для семантичного хешування зображень з використанням гаусівського шуму. Розміри його внутрішніх шарів кодувальника (і декодувальника, в оберненому порядку) становитимуть 512, 256. Розмір кодуючого шару - 20. Використовуватимемо функцію активації ReLU для прихованих шарів, а для вихідного та кодуючого - Sigmoid. Параметр шуму noise_stddev встановимо рівним 0.1. Втрату обчислюватимемо за допомогою Binary Crossentropy. Застосовуватимемо оптимізатор Adam.

```python
class HashAutoEncoderWithNoise:
  def __init__(self, input_size=784, hidden_sizes=[512, 256],
               code_size=20, noise_stddev=0.1, activation="relu"):
    self._input_size = input_size
    self._hidden_sizes = hidden_sizes
    self._code_size = code_size
    self._output_size = self._input_size
    self._noise_stddev = noise_stddev
    self._activation = activation

    inputs = keras.Input(shape=(self._input_size,))

    # Encoder
    x = inputs
    for size in self._hidden_sizes:
      x = keras.layers.Dense(size, activation=self._activation)(x)

    # Gaussian noise
    def noisy_activation(x, training=False):
      if training:
        noise = tf.random.normal(shape=tf.shape(x), mean=0.0,
                                stddev=self._noise_stddev)
        return x + noise
      return x

    pre_code = keras.layers.Dense(self._code_size)(x)
    noisy_code = keras.layers.Lambda(
        lambda x, training: noisy_activation(x, training=training),
        output_shape=(self._code_size,))(pre_code)
    code = keras.layers.Activation("sigmoid")(noisy_code)

    self._encoder = keras.Model(inputs, code)

    # Decoder
    x = code
    for size in reversed(self._hidden_sizes):
      x = keras.layers.Dense(size, activation=self._activation)(x)
    outputs = keras.layers.Dense(self._input_size, activation="sigmoid")(x)

    self._model = keras.Model(inputs, outputs)

  def compile(self, optimizer="adam", loss="binary_crossentropy"):
    self._model.compile(optimizer=optimizer, loss=loss)

  def fit(self, X_train, X_val, epochs=10, batch_size=BATCH_SIZE,
          verbose=1):
    self._model.fit(
        X_train, X_train, epochs=epochs,
        validation_data=(X_val, X_val),
        batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, verbose=1):
    return self._model.evaluate(X_test, X_test, verbose=verbose)

  def encode(self, X_test):
    return self._encoder.predict(X_test)

  def reconstruct(self, X_test):
    return self._model.predict(X_test)

  def encode_hash(self, X_test):
    codes = self.encode(X_test)
    return (codes > 0.5).astype(int)
```

* Вiдобразити зображення, якi мають однаковi хешi. Вiзуально оцiнити чи виглядають однаковими зображення, що мають однаковi хешi. Набори зображень є розмiченими, тому бiльш об’єктивний спосiб вимiрювання ефективностi автокодувальника для семантичного хешування полягає в тому, щоб перевiрити, чи будуть зображення з тим самим хешем мати один i той самий клас.

Створимо функції для групування зображень за хешами та для візуалізації зображень з однаковими хешами.

```python
def group_by_hashes(binary_hashes):
  hash_strings = ["".join(map(str, h)) for h in binary_hashes]

  hash_groups = {}
  for idx, h_str in enumerate(hash_strings):
    if h_str not in hash_groups:
      hash_groups[h_str] = []
    hash_groups[h_str].append(idx)

  return hash_groups

def show_similar_images(group):
  n = len(group)
  plt.figure(figsize=(15, 2))
  for i, idx in enumerate(group):
    plt.subplot(1, n, i+1)
    plt.imshow(X_test[idx].reshape(28, 28), cmap="gray")
    plt.axis("off")
  plt.show()
```

Відобразимо по 10 зображень для 5 груп зображень з однаковими хешами, що мають 10+ зображень.

```python
hash_groups = group_by_hashes(binary_hashes)

# Plot images with the same hash
counter = 0
for h_str, group in hash_groups.items():
  if len(group) >= 10:
    print(f"Hash: {h_str}, Images number: {len(group)}")
    show_similar_images(group[:10])
    counter += 1
  if counter > 5:
    break
```

Бачимо, що зображення з однаковими хем-кодами є дійсно подібними, отже автоенкодер з використанням гусівського шуму для семантичного хешування добре узагальнює.

* Перевiрити, чи будуть зображення з тим самим хешем мати один i той самий клас. Розрахувати середнє значення iндексiв забрудненостi Джинi наборiв зображень з однаковими або дуже схожими хешами. Це середнє значення можна вважати оцiнкою ефективностi автокодувальника для семантичного хешування.

Створимо функції для обчислення індексу Джині для окремого зразка та середнього значення для всього набору.

```python
def gini_impurity(labels):
  total = len(labels)
  if total == 0:
    return 0

  unique_classes, counts = np.unique(labels, return_counts=True)
  probs = counts / total
  return 1 - np.sum(probs ** 2)

def calculate_avg_gini_impurity(hash_groups):
  gini_values = []
  for h_str, indices in hash_groups.items():
    group_labels = y_test[indices]
    gini = gini_impurity(group_labels)
    gini_values.append(gini)
  return np.mean(gini_values)
```

Обчислимо середній індекс Джині для автокодувальника з використанням гаусівського шуму.

```python
avg_gini = calculate_avg_gini_impurity(hash_groups)
print(f"Average Gini Impurity: {avg_gini:.4f}")
```

Низький індекс Джині свідчить про те, що автокодувальник добре групує зображення однакового класу.

Підберемо найкращі параметри для моделі автокодувальника з гаусівським шумом з використанням перехресної перевірки. Розглядатимемо різні значення параметрів code_size (20, 30, 40) та noise_stddevs (0.1, 0.2, 0.3). Для перехресної перевірки використаємо 3 фолди.

```python
code_sizes = [20, 30, 40]
noise_stddevs = [0.1, 0.2, 0.3]
best_loss = float("inf")
best_params = None

kf = KFold(n_splits=3, shuffle=True, random_state=SEED)

for code_size in code_sizes:
  for noise_stddev in noise_stddevs:
    val_losses = []

    for train_index, val_index in kf.split(X_train_mnist):
      X_tr, X_val = X_full_train[train_index], X_full_train[val_index]

      model = HashAutoEncoderWithNoise(
          code_size=code_size, noise_stddev=noise_stddev)
      model.compile()
      model.fit(X_tr, X_val, epochs=5, verbose=0)

      loss = model.evaluate(X_val)
      val_losses.append(loss)

    avg_loss = np.mean(val_losses)
    print(f"Tested code_size={code_size}, noise_stddev={noise_stddev}:",
          f"val MSE: {avg_loss:.4f}")

    if avg_loss < best_loss:
      best_loss = avg_loss
      best_params = (code_size, noise_stddev)

print(f"Best hyperparameters: {best_params} with val loss: {best_loss:.4f}")
```

####vi. Розглянути третю модель - глибоку згорткову мережу для класифiкацiї. Використати вихiдний шар цiєї моделi для отримання хешiв.

Створимо клас глибокої згорткової мережі. Розміри її внутрішніх згорткових шарів становитимуть 32, 64, а глибокого шару - 128. Використовуватимемо функцію активації ReLU для прихованих шарів, для вихідного - Softmax, а для кодуючого - Sigmoid. Розмір кодуючого шару встановимо рівним 20. Втрату обчислюватимемо за допомогою Sparse Categorical Crossentropy. Застосовуватимемо оптимізатор Adam та обчислюватимемо метрику Accuracy під час навчання.

```python
class ConvHashModel:
  def __init__(self, input_shape=(28, 28, 1), conv_sizes=[32, 64],
               dense_size=128, code_size=20, output_size=10,
               activation="relu"):
    self._input_shape = input_shape
    self._conv_sizes = conv_sizes
    self._dense_size = dense_size
    self._code_size = code_size
    self._output_size = output_size
    self._activation = activation
    self._conv_kernel_shape = (3, 3)
    self._maxpool_kernel_shape = (2, 2)

    inputs = keras.Input(shape=self._input_shape)
    x = inputs

    for size in self._conv_sizes:
      x = keras.layers.Conv2D(
          size, self._conv_kernel_shape,
          activation=self._activation)(x)
      x = keras.layers.MaxPooling2D(self._maxpool_kernel_shape)(x)

    x = keras.layers.Flatten()(x)
    x = keras.layers.Dense(
        self._dense_size, activation=self._activation)(x)

    encoded = keras.layers.Dense(self._code_size, activation="sigmoid")(x)
    self._hash_encoder = keras.Model(inputs, encoded)

    outputs = keras.layers.Dense(
        self._output_size, activation="softmax")(x)

    self._model = keras.Model(inputs, outputs)

  def compile(self, optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]):
    self._model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

  def fit(self, X_train, y_train, X_val, y_val, epochs=10,
          batch_size=BATCH_SIZE, verbose=1):
    self._model.fit(
        X_train, y_train, epochs=epochs,
        validation_data=(X_val, y_val),
        batch_size=batch_size, verbose=verbose)

  def evaluate(self, X_test, y_test, verbose=1):
    return self._model.evaluate(X_test, y_test, verbose=verbose)

  def encode_hash(self, X_test):
    codes = self._hash_encoder.predict(X_test)
    return (codes > 0.5).astype(int)
```

Створимо та навчимо згорткову нейроннну мережу. Змінимо розмір вхадних даних на потрібний.

```python
conv_hash_model = ConvHashModel()
conv_hash_model.compile()
conv_hash_model.fit(
    X_train.reshape((-1, 28, 28, 1)), y_train,
    X_valid.reshape((-1, 28, 28, 1)), y_valid)
```
